```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('calib.csv')
data = data[['pi', 'q']]
data
```

```python
def smooth(y, box_pts):
    box = np.ones(box_pts)/box_pts
    y_smooth = np.convolve(y, box, mode='same')
    return y_smooth
```

```python
sort_index = np.argsort(data['pi']).values
plt.scatter(np.arange(160), (smooth(data['pi'][sort_index],4)))
plt.scatter(np.arange(160), (smooth(((data['q'][sort_index]-0.38)*0.3), 2)))
plt.show()
```

```python
sort_index = np.argsort(data['pi']).values
plt.scatter(np.arange(160), (smooth(data['pi'],5)))
plt.scatter(np.arange(160), (smooth(((data['q']-0.38)*0.3), 3)))
plt.show()
```
